```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%pylab inline
from bokeh.core.properties import value
from bokeh.io import show, output_notebook
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, FactorRange
from bokeh.palettes import Spectral6, Spectral8, Category20
from bokeh.transform import factor_cmap

output_notebook()

cust = pd.read_csv('C:/Users/Lara/Documents/Repository/Capstone2/custodial-deaths.csv')
# shoot = pd.read_csv('C:/Users/Lara/Documents/Repository/Capstone2/officer_shootings.csv')
```

```python
deathtype= list(cust['Manner Death'].unique())
```

```python
eth = list(cust['Ethnicity'].unique())
```

```python
breakdown = cust.groupby(['Ethnicity', 'Manner Death'])['Last Name'].count()
death_eth_df = breakdown.unstack()
death_eth_df = death_eth_df.fillna(0).astype(int)
death_eth_df
```

```python
data= {
    'Ethnicity': eth,
    'Accidental injury by others' : list(death_eth_df['Accidental injury caused by others']), 
    'Natural Causes': list(death_eth_df['Natural Causes/Illness']), 
    'Other':list(death_eth_df['Other']), 
    'Accidental injury to self': list(death_eth_df['Accidental injury to self']),  
    'Homicide by Law Enforcement/Correctional Staff': list(death_eth_df['Homicide by Law Enforcement/Correctional Staff']),
    'Suicide': list(death_eth_df['Suicide']),
    'Alcohol/Drug Intoxication': list(death_eth_df['Alcohol/Drug Intoxication']),
    'Homicide by Other': list(death_eth_df['Homicide by Other'])
}

deathtype=['Accidental injury by others',
           'Natural Causes',
           'Other',
           'Accidental injury to self',
           'Homicide by Law Enforcement/Correctional Staff',
           'Suicide',
           'Alcohol/Drug Intoxication',
           'Homicide by Other']

source = ColumnDataSource(data=data)


g = figure(x_range=data['Ethnicity'], plot_height=250, title="Manner of Death by Ethnicity",
           toolbar_location=None, tools="")


g.vbar_stack(deathtype, x='Ethnicity', width=0.9, color=Spectral8, source=source,
             legend=[value(x) for x in deathtype])

g.y_range.start = 0
g.y_range.end = 6000
g.x_range.range_padding = 0.1
g.xgrid.grid_line_color = None
g.axis.minor_tick_line_color = None
g.outline_line_color = None
g.legend.location = "top_left"
g.legend.orientation = "horizontal"

show(g)
```
